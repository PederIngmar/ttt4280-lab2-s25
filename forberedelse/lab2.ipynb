### Oppgave 1



```python
import numpy as np
import matplotlib.pyplot as plt

sampling_f = 1000
t = np.arange(0, 2*np.pi, 1/sampling_f)
x = np.cos(t+1.53)
y = np.sin(t)


plt.plot(t, x, label='cos(t+1.53)')
plt.plot(t, y, label='sin(t)')
plt.legend()
plt.show()
```

```python
abs_cross_correlation = np.correlate(x, y, mode='full')
lags = np.arange(-len(x)+1, len(x))

max_lag = lags[np.argmax(abs_cross_correlation)]
delta_t = max_lag/sampling_f
print(f"Time delay between x and y: {delta_t}")


plt.figure(figsize=(10, 5))
plt.plot(lags, abs_cross_correlation)
plt.title('Cross-Correlation between x and y')
plt.xlabel('Lag')
plt.ylabel('Cross-Correlation')
plt.grid(True)
plt.show()
```
